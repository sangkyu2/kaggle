**나이에 따른 생존률 확인**
================

Sex와 Pclass를 결합하여 예측했더니 score가 더 안좋아졌다. 나이에 따른 생존률 확인용

```python
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv('C:/Users/sangkyu/git/kaggle/titanic/input/train.csv')
test = pd.read_csv('C:/Users/sangkyu/git/kaggle/titanic/input/test.csv')
print(train.shape)
print(test.shape)
```

```python
train.pivot_table('Survived',['Age'],aggfunc=np.mean)
```

```python
test['Survived'] = (test.Age>40)
```

```python
test.head()
```

```python
submission3 = test[['PassengerId','Survived']].copy()
submission3.head()
```

```python
submission3['Survived']=submission3['Survived'].astype(int)
```

```python
submission3.to_csv('../titanic/submission3.csv',index = False)
```
