```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings(action='ignore')
import seaborn as sns
```

```python
train = pd.read_csv('C:/Users/sangkyu/git/kaggle/titanic/input/train.csv')
test = pd.read_csv('C:/Users/sangkyu/git/kaggle/titanic/input/test.csv')
full_data = train.append(test)
```

```python
def titanic_corr(data):
    correlation = data.corr()
    sns.heatmap(correlation, annot=True, cbar=True, cmap="RdYlGn")
    
titanic_corr(train)
# 데이터를 보면 'Sex','PCalss'가 생존이랑 관련된것 같고 'Fare','PClass'는 강한 음의 상관관계를 가지고있다.
```

```python
# Age컬럼 PClass 1,2에는 많이 있고 3에는 별로 없다. -> Age를 가지고 예측하면 noise가 될 확률이 높다.
full_data['Has_Age'] = full_data['Age'].isnull().map(lambda x : 0 if x == True else 1)
fig = plt.figure(figsize=(11,6))
ax = sns.countplot(full_data['Pclass'],hue=full_data['Has_Age'])
ax.set_title('Passenger has age',fontsize = 20)
```

```python
Minor_mask = (train.Age <= 14) 
Not_Minor_mask = (train.Age > 14) 
display(train[Minor_mask][['Pclass','Survived']].groupby(['Pclass'],as_index=False).mean())
display(train[Not_Minor_mask][['Pclass','Survived']].groupby(['Pclass'],as_index=False).mean())
```

```python
# Null값들을 -1로 바꾸고 14세 이하인 사람들을 Minor라는 새로운 feature를 만들어서 저장하였다.
full_data['Age_copy'] = full_data['Age'].fillna(-1)
full_data['Minor'] = (full_data['Age_copy'] < 14.0) & (full_data['Age_copy']>= 0)
full_data['Minor'] = full_data['Minor'] * 1 # 1곱하면 True,False를 숫자형으로 바꿀수 있음
```

```python
train['Minor'] = full_data['Minor'][:len(train)]
test['Minor'] = full_data['Minor'][len(train):]
```
